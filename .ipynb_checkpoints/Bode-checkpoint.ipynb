```python
from pylab import *
```

```python
f = logspace(0,5)
```

```python
w = 2*pi*f
```

```python
H = abs(1j*w)
plot(f, 20*log10(H) ); 
xscale('log')
```

```python
H1 = abs( 1 + (1j*w)/200 )
plot(f, 20*log10(H1) ); 
xscale('log')
```

```python
H =  -20*log10( abs(56250) )
H1 = 20*log10( abs( (1j*w) ) )
H2 = 20*log10( abs( 1 + ((1j*w)/200) ) )
H3 = -40*log10( abs( 1 + ((1j*w)/7500) ) )
H4 = -20*log10( abs( 1 + ((1j*w)*0.05) + (((1j*w)/10 )**2) ))
```

```python
plot(f, (H + H1 + H2 + H3 + H4) ); 
xscale('log')
```
