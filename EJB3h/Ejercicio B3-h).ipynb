Debemos hallar la Transferencia del circuito $H(s) = \frac{V_2(s)}{V_1(s)}$, el tipo de filtro y las respuesta al impulso y al escalón. 


### Tipo de filtro:

Para poder analizar el tipo de filtro sin necesidad de plantear ecuaciones podemos recurrir al análisis del circuito trasnformado mediante la transformada de laplace. Donde la impedancia de un capacitor es $Z_C = \frac{1}{sC}$ y la de un resitor es $Z_R = R$.

#### Frecuencias bajas ($\lim_{s \to 0}$):

En este caso $\lim_{s \to 0} Z_C = \infty$, lo cuál nos dice que los capacitores actuan como un circuito abierto. dando como resultado:

<img src="files/pics/LF.jpeg" width="500" height="40"> 

En la figura se explica como no hay corriente por ninguno de las ramas del circuito y la tensión de salida $V_2 = 0$.

$$\boxed{\lim_{s \to 0}H(s) = 0}$$

#### Frecuencias altas($\lim_{s \to \infty}$):

En este caso $\lim_{s \to \infty} Z_C = 0$, lo cuál nos dice que los capacitores actuan como un corto-circuito dando como resultado:

<img src="files/pics/HF.jpeg" width="500" height="40"> 

Pero este circuito cuando $s \to \infty$ es:

<img src="files/pics/HF_2.jpeg" width="500" height="40"> 


Si vemos la Figura anterior con ganas, es muy similar a una configuración de operacional inversor, solo que en este caso, en vez de resistores, tenemos capacitores. Sin embargo, como estamos trabajando en el circuito transformado podemos utilizar el resultado de la trasnferencia de un invesor.

$$\frac{V_2(s)}{V_1(s)} = - \frac{Z_{C_f}}{Z_C} = - \frac{\frac{1}{sC_f}}{\frac{1}{sC}} = -\frac{C}{C_f}$$

$$\boxed{\lim_{s \to \infty}H(s) = -\frac{C}{C_f}}$$

Vemos entonces que la ganancia en frecuencias altas del circuito depende de la relación entre los valores de estos dos capacitores, de manera que si son iguales $C_f = C$ (como en el enunciado).

$$\boxed{\lim_{s \to \infty} H(s) = -1 }$$

Viendo ambos resultados podemos darnos cuenta que el circuito es un **Filtro Pasa-Altos** o **High-Pass Filter**.

### Transfrencia $H(s)$

Ahora para obtener la expresión analítica de la transferencia planteamos el método de nodos, recordando que no podemos plantear mallas con operacionales ni plantear nodos a la salida de un operacional! Planteamos con el capacitor de realimentación genérico.

<img src="files/pics/nodos.jpeg" width="500" height="40"> 

$$A)~~~~V_1 s C + V_2 s C_f = V_A \left[2sC+sC_f+\frac1R\right] - V_B\left[sC\right]$$
$$B)~~~~\frac{V_2}{R_f} = -V_A \left[sC\right] + V_B\left[sC + \frac{1}{R_f}\right]$$

Luego, como el OP-AMP es ideal, la tensión del nodo B es igual a la del terminal no-inversor del OP-AMP y por ende, $V_B = 0$.

$$A)~~~~V_1 s C + V_2 s C_f = V_A \left[2sC+sC_f+\frac1R\right] $$
$$B)~~~~\frac{V_2}{R_f} = -V_A \left[sC\right] $$

de (B):

$$V_2 = -C R_f \cdot sV_A$$

Si vemos fijo esta ecuación corresponde un circuito integrador, el mismo esta presente en el circuito y corresponde a la siguiente figura, esto nos puede dar idea del tipo de filtro que estamos analizando.

<img src="files/pics/diff.jpeg" width="500" height="40"> 


Despejando de las ecuaciónes A y B obtenemos:

$$H(s) = -\frac{s^2}{s^2 + s \left[\frac{2}{R_f C_f} + \frac{1}{R_fC}\right] + \frac{1}{R_f C_f R C}}$$

De donde podemos comprobar los resultados obtenidos anteriormente y verificar que el filtro es un **pasa-altos**.

$$\lim_{s \to 0} = 0~~~~~~~~\lim_{s \to \infty} = -\frac{C}{C_f}\big|_{C_f=C} = -1$$

Reemplazando por los valores obtenemos la transferencia del circuito:

$$\boxed{H(s) = \frac{V_2(s)}{V_1(s)} = - \frac{s^2}{s^2 + s \frac{1000}{3} + 11111111,11}}$$


### Respuestas al impulso y al escalón

Una vez obtenida la transferencia queremos encontrar la respuesta al impulso y al escalón. Una opción (no recomendable) es simplemente poner $v_1(t) = \delta(t)$ o $v_1(t) = u(t)$ y resolver el circuito como hicimos en la guía 1-B.

La otra opción (recomendada) es pensar lo siguiente: trabajamos con circuitos LTI (Lineales e invariantes en el tiempo). Esto quiere decir que podemos encontrar cualquier salida del circuito a partir de la respuesta al impulso: 

Pues si encontramos una transferencia

$$H(s) = \frac{V_2(s)}{V_1(s)} \implies V_2(s) = V_1(s) H(s)$$

La antitransformada de un producto es la convolución de las antitransformadas:

$$v_2(t) = h(t) \ast v_1(t)$$

Ahora particularmente cuando la entrada es un impulso $v_1(t) = \delta(t)$:

$$v_2(t) = h(t) \ast \delta(t) = h(t)$$

Entonces la respuesta al impulso es $h(t) = \mathcal{L}^{-1} \left\{H(s)\right\}$.

Ahora, si la entrada es un escalón $v_1(t) = u(t)$


$$v_2(t) = h(t) \ast u(t) \overbrace{\implies}^{\mathcal{L}} V_2(s) = H(s) \cdot \frac1s \implies v_2(t) = \mathcal{L}^{-1}\left\{\frac{H(s)}{s}\right\}$$

#### ¿Qué nos esperamos?

Dado que el sistema es de segundo orden vimos que existen tres respuestas posibles que depende de la naturaleza de sus polos (raíces del polinomio característico). En este caso son:

$$p_{1,2} = -\frac{500}{3} \pm 3329,16j$$

De manera que las respuestas que debemos obtener deben tener la pinta de:

$$h(t) = e^{-\frac{500}{3}} \left[k_1 \cos\left(3329,16t\right) + k_2 \sin\left(3329,16t\right)\right] u(t)$$
#### Respuesta al escalón:

$$h_{esc}(t) = \mathcal{L}^{-1}\left\{\frac{H(s)}{s}\right\}$$

$$\frac{H(s)}{s} = - \frac{s}{s^2 + s \frac{1000}{3} + 11111111,11}$$

Recordando la expresión vista en clase del denominador en función de $w_0$ y $Q$:

$$s^2 + s \frac{1000}{3} + 11111111,11 = s^2 + s \frac{w_0}{Q} + w_0^2 $$

De donde obtenemos que $w_0 = \frac{10000}{3}$ y $Q = 10$


Ahora, para antitransformar se propone lo siguiente:

$$s^2 + s \frac{w_0}{Q} + w_0^2 = (s + \frac{w0}{2Q})^2 - \left(\frac{w_0}{2Q}\right)^2 + w_0^2$$



Luego:

$$\frac{H(s)}{s} = - \frac{s}{(s + \frac{w0}{2Q})^2 - \left(\frac{w_0}{2Q}\right)^2 + w_0^2}$$

Para simplificar visualizar las antitransformadas llamamos $w_0^2 - \left(\frac{w_0}{2Q}\right)^2 = a^2$ 
y $k = \frac{w_0}{2Q}$, notar que $a = w_d$ 


$$\frac{H(s)}{s} = - \frac{s}{(s + k)^2 + a^2} = \frac{H(s)}{s} = - \frac{s + k - k}{(s + k)^2 + a^2}$$

$$\frac{H(s)}{s} = - \frac{s + k}{(s + k)^2 + a^2} +  \frac{k}{(s + k)^2 + a^2} $$


$$\frac{H(s)}{s} = - \frac{s + k}{(s + k)^2 + a^2} +  \frac{k}{a}\frac{a}{(s + k)^2 + a^2} $$

Luego, encontramos antitrasnformadas conocidas, el primer termino corresponde a un coseno y el segundo a un seno.

$$h_{esc}(t) = \mathcal{L}^{-1} \left\{\frac{H(s)}{s}\right\} = \left[- e^{-kt} \cos(at) + \frac{k}{a} e^{-kt} sen(at) \right]u(t)$$


La respuesta al escalón es:

$$\boxed{h_{esc}(t) = - e^{-kt} \left[\cos(at) + \frac{k}{a} sen(at)\right] u(t)}$$

$$\boxed{h_{esc}(t) = - e^{-\frac{500}{3}t} \left[\cos(3329,16t) + 0,05 sen(3329,16t)\right] u(t)}$$



#### Respuesta al impulso:

En este caso podríamos antitransformar, pero también recordemos que la respuesta al impulso es la derivada de la respuesta al escalón:

$$h(t) = \frac{\partial h_{esc}(t)}{\partial t}$$

Derivando obtenemos la respuesta al impulso del sistema:

$$\boxed{h(t) = e^{-kt} \left[2k\cos(at) + \left(a- \frac{k^2}{a}\right) sen(at)\right] u(t)}$$

$$\boxed{h(t) = e^{-\frac{500}{3}t} \left[\frac{1000}{3}\cos(3329,16t) + 3320,82\sin(3329,16t)\right] u(t)}$$



```python
w0 = 10000/3
Q = 10

k = w0/(2*Q)
a = (w0 **2 - k ** 2) ** (1/2)

print("k es igual a %.5f"%k)
print("a es igual a %.5f"%a)
print ("2k es igual a %.5f"%(2*k))
print ("a-(k^2/a)es igual a %.5f"%(a-(k**2)/a))
```

```python
import numpy as np
import control as ctrl
import matplotlib.pyplot as plt



s = ctrl.TransferFunction.s
G = -s **2 /(s**2+s*w0/Q+w0 **2);


T = np.arange(0,0.04,1e-6) 

[t,y] = ctrl.step_response(G,T);

h_esc = -np.exp(-k*t)*(np.cos(a*t) + (k/a)* np.sin(a*t))


plt.figure(figsize=[20,10])
plt.plot(t,y)
plt.plot(t,h_esc,'-.',linewidth=2)
plt.grid(True)
plt.title("Respuesta al escalón del circuito")
plt.xlabel("Tiempo(s)")
plt.ylabel(r"$V_2(V)$")



G_bis = G/(s+1e6);

[t,y] = ctrl.impulse_response(G,T);

plt.figure(figsize=[20,10])
plt.plot(t,y)

h = np.exp(-k*t)*(2*k*np.cos(a*t) + (a-(k**2)/a)* np.sin(a*t))

plt.plot(t,h,'-.',linewidth=2)
plt.grid(True)
plt.title("Respuesta al impulso del circuito")
plt.xlabel("Tiempo(s)")
plt.ylabel(r"$V_2(V)$");
plt.legend(["sim","analitica"])
```
