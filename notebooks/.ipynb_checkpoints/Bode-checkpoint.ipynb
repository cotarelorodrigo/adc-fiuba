```python
import control as ctl
import matplotlib.pyplot as plt
import numpy as np
```

```python
G1 = ctl.tf([-1, 0 ,0], [1])
G2 = ctl.tf([1], [1, 1785.71, (3.189)*(10**6)])
```

```python
print(G1)
print(G2)
```

```python
ctl.bode(G1, dB=True, label='G1')
ctl.bode(G2, dB=True, label='G2')
plt.legend()
plt.figure(figsize=(500,500))
plt.show()
```

```python
ctl.bode(G1*G2, dB=True)
plt.figure(figsize=(500,500))
plt.show()
```
